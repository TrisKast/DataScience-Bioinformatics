## Task: Plot the distribution patterns of maternal and paternal alleles across strains and SNPs

### Step 1: convert cell values to numbers

Note on heterozygous loci: "It is recommended to treat these rare heterozygous loci as missing data and ensure that only the additive effects of B vs. D alleles are estimated" (http://www.genenetwork.org/webqtl/main.py?FormID=sharinginfo&GN_AccessionId=600)

Hint: For the Heatmap in the next task, it is preferable to have a [-x,x] range, rather than a [0,x] range.

```python
encode = {'B' : -1, 'D' : 1, 'U' : 0, 'H': 0}

#Using pandas applymap function to make it more smooth than with for loops
df_geno_numerical = df_geno.applymap(encode.get)
df_geno_numerical.head()
```

### Step 2: Plot clustered heatmap

```python
cmap = sns.clustermap(df_geno_numerical, figsize=(10,10))
```

This clustered heatmap does hierarchical clustering on the mouse dataset. Blocks of black show related mouse but since its rather noisy the mouse were apparently well mixed. Individuals that are closer in the dendrogram show rather similar distributions of black.

### Plot distribution of maternal and paternal alleles

Hint: Count occurances of the alleles B and D for each strain

The comparison of the distributions can be done with boxplots or with a histogram. Both can be seen below.

```python
#Use a Counter object to count the occurences of -1's, 0's, and 1's in the table

df = []
for col in df_geno_numerical.columns:
    df.append(Counter(df_geno_numerical[col]))

#Transform the list into a dataframe
df_occ = pd.DataFrame.from_dict(df, orient='columns')
df_occ.columns = ['B', 'U/H', 'D']     
```

```python
#Make boxplot
fig, ax = plt.subplots(figsize=(10,8))
sns.set(context='notebook', style='whitegrid')    
bplot = sns.boxplot(data=df_occ)  

plt.title('Boxplot of the distribution of maternal,paternal and hybrid/unknown alleles', fontsize= 15)
plt.xlabel('Distribution', fontsize= 15)

plt.show()
```

One can nicely see the almost equal distributions of maternal and paternal allels (green & blue boxes) and the very low numner of unknown and hybrid alles (green). The last one is also shiftet to zero, since all counts for "unknown" are labeled as 0.

```python
#Also make a bar plot
df_occ.fillna(value=0, inplace=True)
df_occ_trans = df_occ.transpose()

fig, ax = plt.subplots(figsize=(10,8))

hplot = plt.hist(df_occ[df_occ.columns[0]], alpha = 0.5, label = 'B')
hplot = plt.hist(df_occ[df_occ.columns[2]], alpha = 0.5, label = 'D')

plt.xlabel('Values', fontsize= 15)
plt.ylabel('Value counts', fontsize= 15)
plt.title('Histogram of the paternal and maternal allel distribution', fontsize= 15)

plt.legend()
plt.show()
```

Again and as expected the distribution of maternal and paternal SNP heritance is roughly the same. For the histogram i skipped the hybrid/unknown data.

### Subtask: Compare the two different diets (for several biochemistry measurements)
Tip: after prototyping your analysis, wrap your individual steps in a function that you can use for an arbitrary phenotype in an arbitrary worksheet

1. extract columns we are interested in from full dataframe
2. add a column with the difference between HFD and CD
3. plot all three columns for all strains as a bar plot (see e.g. pandas.DataFrame.plot.bar )

```python
#Read in some phenotype and extract all columns which start with HFD and CD
def readSheet(sheetName):
    dataset = pd.read_excel(xls, sheet_name=sheetName, na_values='x', index_col='@format=column' )
    dataset = dataset.loc[:,(dataset.columns.str.startswith("HFD")) | (dataset.columns.str.startswith("CD"))]
    dataset = dataset.fillna(0)
    return dataset
```

```python
#1. extract columns we are interested in from full dataframe
dataset_bc = readSheet('Biochemistry')
dataset_bc_filtered = dataset_bc.loc[:,(dataset_bc.columns.str.startswith('CD_Cholesterol_[mmol/L]')) | (dataset_bc.columns.str.startswith('HFD_Cholesterol_[mmol/L]'))]
dataset_bc_filtered.head()
```

```python
#2. add a column with the difference between HFD and CD
substracts = dataset_bc_filtered['CD_Cholesterol_[mmol/L]'].sub(dataset_bc_filtered['HFD_Cholesterol_[mmol/L]']) 

dataset_bc_filtered['differences'] = substracts
dataset_bc_filtered.head()
```

```python
#3. plot all three columns for all strains as a bar plot

fig, ax = plt.subplots(figsize=(8,6))
bplot = sns.barplot(data=dataset_bc_filtered)

plt.title('Differences in Cholesterol [mmol/L] levels for the two diets across strains', fontsize= 15)

plt.show()
```

The comparison between the cholesterol level between the mice on CD and HFD diet. The level is strongly increased for the mice with the highh fat diet

```python
#Alternatively as stacked bar plot
sbplot = dataset_bc_filtered.plot.bar(stacked=True, figsize=(12,6))
plt.title('Differences in Cholesterol [mmol/L] levels for the two diets across strains', fontsize= 15)

plt.show()
```

The advantage of this stacked bar plot is that one can easily detect the strains with the highest difference in the compared attribute by simply looking at the red bars. Butin my opinion the overall difference of all strains is more interesting, so I included the normal bar plot in the following function.

__If you did not do it already, define functions that perform aspects of the procedure above__

```python
#Function to compare two chooseable measurements in a chooseable phenotype
def compare_phenotype_attributes(phenotype, measurement1, measurement2):
    #First read in the phenotype
    dataset = readSheet(phenotype)
    #Then select the two interesting attribteus
    dataset_filtered = dataset.loc[:,(dataset.columns.str.startswith(measurement1)) | (dataset.columns.str.startswith(measurement2))]
    
    #Substract them from each other and save the differences in a new column
    substracts = dataset_filtered[measurement1].sub(dataset_filtered[measurement2]) 
    substracts.abs()
    dataset_filtered['differences'] = substracts
    
    #Plot the differences in a bar plot
    fig, ax = plt.subplots(figsize=(8,6))
    bplot = sns.barplot(data=dataset_filtered)
    plt.title('Differences in '+ measurement1 +' and '+ measurement2 +' for the '+ phenotype +' phenotype', fontsize= 15)


    plt.show()
```

__After refactoring code into functions: Apply function to other phenotypes__

```python
readSheet('Biochemistry').head()
```

The comparison between the glucose level between the mice on CD and HFD diet. The levels are very similiar, lightly increased for the CD mice.

```python
readSheet('Hematology').head()
```

```python
compare_phenotype_attributes('Hematology', 'CD_RBC.TotalRedBloodCells_[#*10^6/uL]', 'HFD_RBC.TotalRedBloodCells_[#*10^6/uL]')
```

## Study Linkage Disequilibrium between the SNPs on each chromosome

__Linkage disequilibrium__ (LD) is the non-random association of alleles at multiple loci. If, in a population, all loci on a chromosome where completely independently distributed, the population would be considered to be at linkage equilibrium - a state that only theoretically reached in population genetics. Stemming from a single founder pair (in complete LD), each generation of offspring observed a set of recombination and crossover events breaking up the structure of the chromosome. This does not necessarily happen at gene boundaries, resulting in set of genes usually inherited together. The state of linkage equilibrium is similar to the Hardy-Weinberg equilibrium that you should have heard about in the context of genotype frequency. In the case of LD we work with haplotypes, however. While the genotype describes the allelic status of a single locus on both chromosomes (A/a), a haplotype expands this scenario to two or more loci that are inherited together from a single parent. If we have two loci with two possible alleles A/a and B/b the following haplotypes can exist

|   | B  | b  |
|---|----|----|
| A | AB | Ab |
| a | aB | ab |

In a population the frequency of each of these haplotypes can be computed from the data by counting the number of occurrences of the haplotype and dividing it by the number of samples. In our case we can thus compute the frequency for observing the haplotype and $p_{AB} = n_{AB}/n$ (and the rest analogously)

| Haplotype | A-B      | A-b      | a-B      | a-b      |
|-----------|----------|----------|----------|----------|
| Count     | $n_{AB}$ | $n_{Ab}$ | $n_{aB}$ | $n_{ab}$ |
| Frequency | $p_{AB}$ | $p_{Ab}$ | $p_{aB}$ | $p_{ab}$ |

From a statistical point of view, LD can be calculated as the difference of haplotype frequencies of the alleles at two loci observed in the population compared to the expected haplotype frequency if the two loci where independent $p_A$ and $p_B$. If the coefficient of linkage disequilibrium D between these probabilities is not zero, the two loci are in not independent, i.e. $p_{AB} \ne p_{A}p_{B}$.

$$D = p_{AB} - p_{A}p_{B}$$

Usually, this measure is normalized by dividing by the theoretical maximum for the observed allele frequencies

$$D' = D / D_\text{max}$$

with

$$D_\text{max} =
\begin{cases}
\min\{p_{A}p_{B}, (1-p_{A})(1-p_{B})\} &\mbox{if } D < 0,\\
\min\{(p_{A}(1-p_{B}), (1-p_{A})p_{B}\} &\mbox{if } D > 0
\end{cases}
$$

You can read up more about the concept in the following review:
[Montgomery Slatkin, Linkage disequilibrium — understanding the evolutionary past and mapping the medical future](http://www.nature.com/nrg/journal/v9/n6/full/nrg2361.html). A nice presentation about the intuition can also be found [here](http://csg.sph.umich.edu/abecasis/class/666.03.pdf).
