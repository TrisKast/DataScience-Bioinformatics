# Create a DataFrame object, access its rows, columns, indices, cells

Create a dataframe of the following table.

|index   |a |b  |c   |d          |e |
|--------|--|---|----|-----------|--|
|1       |2 |3  |5.  |horse      |13|
|2       |3 |1  |1.4 |cat        |11|
|3       |4 |9  |1.2 |dog        |  |
|4       |5 |12 |0.8 |tortoise   |  |
|5       |6 |2  |2.3 |rabbit     |9 |

Now
* access the values of each row individually
* access single cells
* print the index and column names
* access values of each column individually
* rename columns
* extract rows 2-4 as new dataframe
* obtain summary table of dataframe

```python
df = pd.DataFrame({
                   'a' : [2,3,4,5,6],
                   'b' : [3,1,9,12,2],
                   'c' : [5.,1.4,1.2,0.8,2.3],
                   'd' : ["horse","cat","dog","tortoise","rabbit"],
                'e' : [13,11, None, None,9]}, index=[1,2,3,4,5])
df
```

```python
#acess individuall row with iloc

df.iloc[1]
```

```python
#access individual cell

df.iloc[2,3]
#OR
#print (df.loc[3, 'd'])
```

```python
#print the index and column names

print(df.index.values, df.columns.values)
```

```python
#access specific column by column name

df['a']
```

```python
#rename colum
df_renamed = df.rename(columns={'a': 'A', 'b': 'B', 'c': 'C', 'd': 'D', 'e': 'E'})
df_renamed
```

```python
#extract rows 2-4 as new dataframe
df2 = df.iloc[1:4] 
df2
```

```python
#extract columns
df.loc[:,'b':'d']
```

```python
#Summary table
df.describe()
```

# Data Merging

* Merge df3 with df4 on columns c, d and e. Change columns with same names in a meaningful way.

```python
#merge with inner join

pd.merge(df3, df4, how='inner', on=['c', 'd', 'e'])
```

* Now perform a full outer join of the data, to keep all rows contained in either of the two dataframes.

```python
#merge outer join

pd.merge(df3, df4, how='outer', on=['c', 'd', 'e'])
```
